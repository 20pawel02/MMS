```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t
from sklearn.linear_model import LinearRegression
```

```python
dane = []
with open("lab8_dane.txt", 'r') as fi:
    for line in fi:
        if line.split():
            line = [float(x) for x in line.split()]
            dane.append(line)


X = []
Y = []
for item in dane:
        X.append(item[1:])
        Y.append(item[0])
print(X)
print(Y)
```

```python
# model liniowy korelacji wielorakiej

model = LinearRegression().fit(X, Y)
r_sq = model.score(X, Y)
print(f"współczynnik determinacji: {r_sq}")
print(f"wyraz wolny: {model.intercept_}")
print(f"współczynniki modelu: {model.coef_}")
print(f"model liniowy: y = {round(model.intercept_ , 2)} + {round(model.coef_[0], 2)}*x1 + {round(model.coef_[1],2)}*x2")
```

```python
dane = []
with open("zachorowania.txt", 'r') as fi:
    for line in fi:
        if line.split():
            line = [float(x) for x in line.split()]
            dane.append(line)


X = []
Y = []
for item in dane:
        X.append(item[0])
        Y.append(item[1])
print(X)
print(Y)
```

```python
# model wielomianowy
X_wiel = [[x, x**2] for x in X]
model = LinearRegression().fit(X_wiel, Y)
r_sq = model.score(X_wiel, Y)
print(f"współczynnik determinacji: {r_sq}")
print(f"współczynnik przesunięcia: {model.intercept_}")
print(f"współczynnik regresji: {model.coef_}")
a = model.coef_
b = model.intercept_
print(f"model - wielomian stopnia dwa:  y = {round(b, 2)} + {round(a[0],2)}*x + {round(a[1], 2)}* x **2 ")
x = 36
prognoza_x = b + a[0]*x + a[1]* x**2 
print(f"prognoza_x: {prognoza_x}")


```
